```python
from garch_models import *
```

```python
stock = 'SPY'
start = '2010-01-01'
end = '2022-08-01'

data = yf.download(stock, start, end, progress=False)
data.drop(['Open', 'High', 'Low', 'Close', 'Volume'], axis='columns', inplace=True)
ret = data['Adj Close'].pct_change().dropna()*100
```

```python
tests = garch_models.tests(ret)
mod1 = garch_models.garch(ret)
mod2 = garch_models.garch(ret, dist='student_t')
mod3 = garch_models.gjr_garch(ret)
mod4= garch_models.gjr_garch(ret, dist='student_t')
mod5 = garch_models.egarch(ret)
mod6 = garch_models.egarch(ret, dist='student_t')
```

```python
tests.jarque_bera()
```
