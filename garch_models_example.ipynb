```python
from garch_models import *
```

```python
stock = 'SPY'
start = '2018-01-01'
end = '2022-08-01'

data = yf.download(stock, start, end, progress=False)
data.drop(['Open', 'High', 'Low', 'Close', 'Volume'], axis='columns', inplace=True)
ret = data['Adj Close'].pct_change().dropna()*100
```

```python
mod1 = garch_models.garch(ret)
mod2 = garch_models.gjr_garch(ret)
mod3 = garch_models.egarch(ret)
```

# GARCH

```python
mod1.plot(), mod1.coef();
```

# GJR-GARCH

```python
mod2.plot(), mod2.coef();
```

# EGARCH

```python
mod3.plot(), mod3.coef();
```
